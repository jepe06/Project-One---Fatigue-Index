```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('results.xlsx')
df.sample(2)
```

```python
result = df.groupby('data_file').agg(['mean', 'std']).reset_index()
result
```

```python
# Step 1: Filter columns by level
df = result.copy()
mean_columns = df.xs('mean', level=1, axis=1)
std_columns = df.xs('std', level=1, axis=1)

# Step 2: Calculate mean of means and mean of stds for each row
df[('overall', 'mean_of_means')] = mean_columns.mean(axis=1)
df[('overall', 'mean_of_stds')] = std_columns.mean(axis=1)

df.set_index('data_file', inplace=True)

df = df[[df.columns[-2]] + [df.columns[-1]] + list(df.columns[:-2])]

# Insert a blank column between the first and second columns
df.insert(2, '', '')

df
```

```python
df.to_excel('results_summary.xlsx', index=True)
```
